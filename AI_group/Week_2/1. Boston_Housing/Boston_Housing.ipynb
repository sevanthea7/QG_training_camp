6. 住宅平均房间数：平均房间数与平均房价呈正线性相关，房间数越多，平均房价越高。

9. 到高速公路的便利程度：到高速公路不便利的房价基本较低，房价高的地带到高速公路基本都很便利。
10. 每$10k的财产税率：到财产税率高的地的房价基本较低，房价高的地带财产税基本都较低。

构造划分训练集和测试集的函数

```python
def preprocess_data(X, y, test_size=0.2, random_state=1000):
    # 随机排列
    np.random.seed(random_state)
    if X.ndim == 1:
        X = np.array(X).reshape(-1, 1)
    X = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)
    indices = np.random.permutation(X.shape[0])
    # 划分训练集测试集
    test_size = int(X.shape[0] * test_size)
    X_train, X_test = X[indices[:-test_size]], X[indices[-test_size:]]
    y_train, y_test = y[indices[:-test_size]], y[indices[-test_size:]]
    return X_train, X_test, y_train, y_test
```

构造线性回归预测的类

```python
class LinearRegression:
    def __init__(self):
        self.mdl_ = None

    def fit_analytical(self, X, y):
        #  最小二乘法
        self.mdl_ = np.linalg.inv(X.T @ X) @ X.T @ y

    def fit_gradient_descent(self, X, y, learning_rate=0.01, n_iterations=5000):
        # 梯度下降
        n_samples, n_features = X.shape
        self.mdl_ = np.zeros(n_features)
        for k in range(n_iterations):
            # 梯度裁剪
            clip_threshold = 0.5
            gradients = -(2 / n_samples) * X.T @ (y - X @ self.mdl_)
            if clip_threshold:
               gradients = np.clip(gradients, -clip_threshold, clip_threshold)
            self.mdl_ -= learning_rate * gradients

    def predict(self, X):
        return X @ self.mdl_
```

低收入人口比例预测

```python
X_LSTAT, y = data['LSTAT'], data['MEDV']

# 划分训练集、测试集
X_train_L, X_test_L, y_train_L, y_test_L = preprocess_data(X_LSTAT, y)

# 实例化
analytical_L = LinearRegression()
gradient_descent_L = LinearRegression()

# 模型训练
analytical_L.fit_analytical(X_train_L, y_train_L)
gradient_descent_L.fit_gradient_descent(X_train_L, y_train_L)

# 在测试集上进行预测
y_predict_analytical_L = analytical_L.predict(X_test_L)
y_predict_gradient_descent_L = gradient_descent_L.predict(X_test_L)

# 计算MSE
mse_LR_L = np.mean((y_test_L - y_predict_analytical_L) ** 2)
mse_SGD_L= np.mean((y_test_L - y_predict_gradient_descent_L) ** 2)

mse_LR_L, mse_SGD_L
```
